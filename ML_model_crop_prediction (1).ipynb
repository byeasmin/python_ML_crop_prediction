```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("field_crop_data.csv")
```

```python
data.head() #to show first 5 rows & coloumns
```

```python
data.shape # to show the dimension
```

```python
data.isnull().sum() # to check if there is any null value or not
```

**Split Features & Labels**

```python
x = data.iloc[:, : -1] # this is features,
                       # : - 1 to 8 coloumns, :-1 - 1 to 7 coloumns

y = data.iloc[:, -1] # this is label,
                     # :,-1 - 8 number coloumns
```

**Training & Testing Data**

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

**Model**

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

**Accuracy**

```python
accuracy = model.score(X_test, y_test)
```

```python
print("Accuracy:", accuracy)
```

**Make Predictions**

```python
new_features = [[5, 47, 21, 93.66024, 60.31891, 7.399136, 116.9263]]
predicted_crop = model.predict(new_features)
print("Predicted crop : ", predicted_crop)
```
